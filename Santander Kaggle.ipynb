```python
#Understand what PCA is originally meant for :-

#Given a very high dimensional input data, that is known to be highly correlated (with well-known correlation patterns 
#that can be described in human-understandable terms), try to extract one or more linear combinations ("Principal 
#Components") from the input data to achieve "decorrelation", with dimensionality reduction as an added benefit.
```

```python
pca = decomposition.PCA(n_components=14)
pca.fit(santander_df)
afterPCA = pca.transform(santander_df)
print(pca.explained_variance_)
#We can see like 14 components are important after not mentioning n_component parameter

#Reference - http://scikit-learn.org/dev/tutorial/statistical_inference/unsupervised_learning.html#principal-component-analysis-pca
#Choosing components
```

# Programatically finding the optimum number of features to select for better accuracy with naive bayes

```python

best_score = 0

number_of_components_list = []
scores_list = []

for number_of_components in range(2,370):
    pca = decomposition.PCA(n_components=number_of_components)
    pca.fit(santander_df)
    afterPCA = pca.transform(santander_df)

    santander_df_after_PCA = pd.DataFrame(afterPCA)
    santander_df_after_PCA.index = santander_df.index
    
    Xtrain, Xtest, ytrain, ytest = train_test_split(santander_df_after_PCA,classLabel,test_size=0.2,random_state=42)
    
    clf = BernoulliNB().fit(Xtrain, ytrain)
    score = clf.score(Xtest,ytest)

    number_of_components_list.append(number_of_components)
    scores_list.append(score)
    
    print("Score= %f , n_components = %f" % (score, number_of_components))
    if(score >= best_score):
        best_score = score
        best_clf = clf
```

```python
import matplotlib.pyplot as plt
plt.plot(number_of_components_list, scores_list, 'ro')
plt.axis([0, 371, 0, 1])
plt.show()
```

# Performing above thing with k fold cross validation, instead of train_test_split

```python
best_score_kfold_NB = 0

number_of_components_list_kfold_NB = []
scores_list_kfold_NB = []
mean_squared_error_list_kfold_NB = []
for number_of_components in range(2,370):
    pca = decomposition.PCA(n_components=number_of_components)
    pca.fit(santander_df)
    afterPCA = pca.transform(santander_df)

    santander_df_after_PCA = pd.DataFrame(afterPCA)
    santander_df_after_PCA.index = santander_df.index
    
    kf = KFold(len(santander_df_after_PCA.axes[0]), n_folds=10)
    for train, test in kf:        
        Xtrain, Xtest, ytrain, ytest = santander_df_after_PCA.iloc[train], santander_df_after_PCA.iloc[test], classLabel[train], classLabel[test]
    
    clf = BernoulliNB().fit(Xtrain, ytrain)
    
    y_predict = pd.DataFrame(clf.predict(Xtest))
    y_predict.index = ytest.index
    mse = mean_squared_error(ytest, y_predict)
    
    score = clf.score(Xtest,ytest)

    number_of_components_list_kfold_NB.append(number_of_components)
    scores_list_kfold_NB.append(score)
    mean_squared_error_list_kfold_NB.append(mse)
    
    print("Score= %f , n_components = %d, mse=%f" % (score, number_of_components,mse))
    if(score >= best_score_kfold_NB):
        best_score_kfold_NB = score
        best_clf_kfold_NB = clf
```

```python
import matplotlib.pyplot as plt
plt.plot(number_of_components_list_kfold_NB, mean_squared_error_list_kfold_NB, 'ro')
plt.axis()
plt.show()
```
