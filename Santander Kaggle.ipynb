```python
#Popping target variable before PCA fit.
classLabel = santander_df.pop('TARGET')
```

```python
#Understand what PCA is originally meant for :-

#Given a very high dimensional input data, that is known to be highly correlated (with well-known correlation patterns 
#that can be described in human-understandable terms), try to extract one or more linear combinations ("Principal 
#Components") from the input data to achieve "decorrelation", with dimensionality reduction as an added benefit.
```

```python
pca = decomposition.PCA(n_components=14)
pca.fit(santander_df)
afterPCA = pca.transform(santander_df)
print(pca.explained_variance_)
#We can see like 14 components are important after not mentioning n_component parameter

#Reference - http://scikit-learn.org/dev/tutorial/statistical_inference/unsupervised_learning.html#principal-component-analysis-pca
#Choosing components
```

```python
santander_df_after_PCA = pd.DataFrame(afterPCA)
santander_df_after_PCA.index = santander_df.index
```

```python
santander_df_after_PCA
```

```python
# Naive Bayes implementation


print ("Number of instances and number of attributes in Santander dataset are %d and %d" % (len(santander_df_after_PCA.axes[0]),len(santander_df_after_PCA.axes[1])))
f = open('ReportOfRunningScript.txt','w')
f.write("Number of instances and number of attributes in Santander dataset are %d and %d" % (len(santander_df_after_PCA.axes[0]),len(santander_df_after_PCA.axes[1])))
f.close()

# Variable to initialize random_state attribute to select different train test split each time. Used in splitting inside the loop.
random_sample_selection_number = [42, 17]
Xtrain, Xtest, ytrain, ytest = train_test_split(santander_df_after_PCA,classLabel,test_size=0.1,random_state=random_sample_selection_number.pop())

```

```python
# Fitting the classifier
clf = BernoulliNB().fit(Xtrain, ytrain)
score = clf.score(Xtest,ytest)
print("Accuracy of Santander dataset after running with Naive Bayes algorithm is %f" % (score))
f = open('ReportOfRunningScript.txt','a')
f.write("Accuracy of Santander dataset after running with Naive Bayes algorithm is %f" % (score))
f.close()
```
